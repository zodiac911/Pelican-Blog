6.使用 ?: 运算符

```cpp
//Yes
return a<b ? a:b;
//No
if (a < b)
    return a;
else
    return b;
```

6.定义函数时尽可能使用按引用传递参数，而不是按值传递参数，能加 const 尽量加上

```cpp
//Yes
int sum (int& a, int& b){
    return a+b;
}

//No
int sum (int& a, int& b){
    return a+b;
}
//Update
//返回引用的函数，return 的必须为左值
int& sum(int& a, int& b) {
	int c = a + b;
	return c;
}
```
